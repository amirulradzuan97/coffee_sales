```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
coffee_sales = pd.read_csv(r"C:\Users\ADMIN\Desktop\Amirul\Data Analyst Files\Project Portfolio\Coffee shop sales\Coffee Shop Sales.csv")

coffee_sales 
```

```python
#total sales amount 
total_sales_amount = coffee_sales['unit_price'].sum()

total_sales_amount
```

```python
#total transaction count
total_sales_transaction = coffee_sales['transaction_id'].count()

total_sales_transaction
```

```python
#total sales by store location
total_sales_by_store = coffee_sales.groupby('store_location')['unit_price'].sum()

total_sales_by_store 
```

```python
#plotting the sales by store

plt.figure(figsize=(10,5))

total_sales_by_store.plot(kind='bar') 

plt.title('Total sales by store')

plt.xlabel('Store')

plt.ylabel('Total Sales')

plt.show()
```

```python
#total sales by product category
total_sales_by_product = coffee_sales.groupby('product_category')['unit_price'].sum()

total_sales_by_product 
```

```python
#plotting the sales by product category

plt.figure(figsize=(10,5))

total_sales_by_product.plot(kind='barh') 

plt.title('Total sales by product category')

plt.xlabel('Total sales')

plt.ylabel('Product category')

plt.show()
```

```python
total_sales_by_product_type = coffee_sales.groupby('product_type')['unit_price'].sum()

total_sales_by_product_type 
```

```python
#plotting the sales by product category

plt.figure(figsize=(12,6))

total_sales_by_product_type.plot(kind='barh') 

plt.title('Total sales by product type')

plt.xlabel('Total sales')

plt.ylabel('Product type')

plt.show()
```
